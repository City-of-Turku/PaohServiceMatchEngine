```python
import numpy as np
import pandas as pd
import json
import pickle
```

```python
with open('services_data.json', 'r') as f:
    services = json.load(f)
with open('service_vector_data.pkl', 'rb') as f:
    service_vectors = pickle.load(f)
```

```python
def get_service(service_id):
    filtered_services = [
        service for service in services if service["id"] == service_id]
    if len(filtered_services) > 0:
        return(filtered_services[0])
    else:
        return(None)
```

```python
len(service_vectors)
```

```python
for vector in service_vectors:
    vector['name'] = get_service(vector['id'])['name']['fi']
```

```python
from whatlies.embedding import Embedding
from whatlies.embeddingset import EmbeddingSet
from whatlies.transformers import Pca, Umap
```

```python
names = [vector['name'] for vector in service_vectors]
vecs = [vector['vector'] for vector in service_vectors]
```

```python
emb = EmbeddingSet.from_names_X(names, vecs)
```

```python
#emb_pca = emb.transform(Pca(2))
emb_umap = emb.transform(Umap(2))
```

```python
emb_umap.plot_interactive().properties(height=1200, width=1200)
```
